```python
brackley_s['codon'] = brackley_s.index.to_series().map(brackley_dict)
```

```python
brackley_s['codon_0'] = brackley_s['codon'].map(lambda x: x[0].lower())
brackley_s['codon_1'] = brackley_s['codon'].map(lambda x: x[1].lower() if len(x) > 1 else '')
```

```python
codon_anticodon = {
    'uuu': 'gaa', 'ucu': 'aga', 'uau': 'gua', 'ugu': 'gca',
    'uuc': 'gaa', 'ucc': 'aga', 'uac': 'gua', 'ugc': 'gca',
    'uua': 'uaa', 'uca': 'uga', 'uaa': '*', 'uga': '*',  # '*'==stop
    'uug': 'caa', 'ucg': 'cga', 'uag': '*', 'ugg': 'cca',
    'cuu': 'gag', 'ccu': 'agg', 'cau': 'gug', 'cgu': 'acg',
    'cuc': 'gag', 'ccc': 'agg', 'cac': 'gug', 'cgc': 'acg',
    'cua': 'uag', 'cca': 'ugg', 'caa': 'uug', 'cga': 'acg',
    'cug': 'uag', 'ccg': 'ugg', 'cag': 'cug', 'cgg': 'ccg',
    'auu': 'aau', 'acu': 'agu', 'aau': 'guu', 'agu': 'gcu',
    'auc': 'aau', 'acc': 'agu', 'aac': 'guu', 'agc': 'gcu',
    'aua': 'uau', 'aca': 'ugu', 'aaa': 'uuu', 'aga': 'ucu',
    'aug': 'cau', 'acg': 'cgu', 'aag': 'cuu', 'agg': 'ccu',
    'guu': 'aac', 'gcu': 'agc', 'gau': 'guc', 'ggu': 'gcc',
    'guc': 'aac', 'gcc': 'agc', 'gac': 'guc', 'ggc': 'gcc',
    'gua': 'uac', 'gca': 'ugc', 'gaa': 'uuc', 'gga': 'ucc',
    'gug': 'cac', 'gcg': 'ugc', 'gag': 'cuc', 'ggg': 'ccc'
}
```

```python
brackley_s['anticodon_0'] = brackley_s['codon_0'].map(codon_anticodon)
brackley_s['anticodon_1'] = brackley_s['codon_1'].map(codon_anticodon)
```

It worked: where 2 codons were available, they mapped to the same anticodon. We can therefore safely use anticodon_0.

```python
brackley_s = pd.merge(brackley_s, peak_frequency, left_on='anticodon_0', right_on='anticodon', how='outer')
```

```python
brackley_s['codon_time'] = brackley_s['codon_speed']**-1
```

```python
myedgecolor = 'steelblue' # powderblue' # 
myfacecolor = 'cornflowerblue' #
```

```python
ax = brackley_s.plot(kind='scatter', x='peak_frequency', y='codon_speed', figsize=(10, 7), edgecolor=myedgecolor, c=myfacecolor, s=40, alpha=1.0)
ax.set_xlabel('Peak frequency per anticodon occurrence')
ax.set_ylabel('Codon speed [relative units]')
ax.grid(True, linestyle = '-', color = '0.75')
plt.savefig('codon_speed_vs_peak_frequency.pgf', bbox_inches='tight', dpi=500)
```
